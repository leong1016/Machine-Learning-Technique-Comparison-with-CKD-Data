```python
import pandas as pd
```

```python
training = pd.read_csv('train.csv')
testing = pd.read_csv('test.csv')
```

```python
whole = pd.concat([training, testing])
whole.head()
```

```python
result = whole.sample(frac=1).reset_index(drop=True)
result.head()
```

```python
result.to_csv('ckd.csv', index=None)
```
